```python
df1 = pd.read_excel("../practise/data/coalpublic2013.xls", header =3,index_col=1)

df1.head()
```

```python
df1['Company Type'].unique()
```

```python
df1['Company Type'].replace(to_replace='Indepedent Producer Operator', value='Independent Producer Operator',inplace=True)

df1['Company Type'].head()
```

```python
df1.rename(columns=lambda x: x.replace(" ", "_"),inplace=True)

df1.head()
```

```python
len(df1)
```

```python
# Relationship between Average_Employees and Labour_hours using a scatter plot
# and a regression plot

plt.scatter(df1.Average_Employees, df1.Labor_Hours)
plt.xlabel('Number of Employees')
plt.ylabel('Total Hours Worked')
```

```python
sns.regplot(df1.Average_Employees, df1.Labor_Hours)

# save figure as image
plt.savefig("../practise/data/2023-4-26_employess_vs_hours.png")
```

```python
# Relationship between Labour_Hours and Production_(short_tones)

plt.scatter(df1.Labor_Hours, df1['Production_(short_tons)'])
plt.xlabel('Hours')
plt.ylabel('Tons')
```

```python
df1['Production_(short_tons)'].hist()
```

```python
# check for production tons with zero production 
df1[df1['Production_(short_tons)'] == 0]
```

```python
# check production_tons above zero
df1[df1['Production_(short_tons)'] > 0]
```

```python
# romove all zero production tons
# data has been removed here#

df1 = df1[ df1['Production_(short_tons)']> 0]
```

```python
# create log_production column to log production tons column using numpy

df1['Log_Production'] = np.log(df1['Production_(short_tons)'])
```
